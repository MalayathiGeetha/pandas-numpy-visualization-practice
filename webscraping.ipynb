```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
req=requests.get("https://www.python.org/")
soup=BeautifulSoup(req.content,"html.parser")
print(soup.get_text())
res=soup.title
print(res.prettify())
print(soup.prettify())
print(soup.title.string)
print(soup.title.parent.name)
for link in soup.find_all('a'):
    print(link.get('href'))
print(soup.get_text())
div=soup.div #gives about the first div tag
print(div)
```
