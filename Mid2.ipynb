```python
import matplotlib.pyplot as plt
import numpy as np

# Data
stocks = ['S1', 'S2', 'S3', 'S4', 'S5']
investments = [25, 15, 18, 20, 10]  # in Crore
returns = [10, 15, 21, 15, 10]  # in Percentage

# Bar Chart - Side by Side
x = np.arange(len(stocks))
width = 0.4

plt.figure(figsize=(10, 5))
plt.bar(x - width/2, investments, width, label="Investments (Crore)", color='b')
plt.bar(x + width/2, returns, width, label="Returns (%)", color='g')

plt.xticks(x, stocks)
plt.xlabel("Stock Companies")
plt.ylabel("Values")
plt.title("Stock Investments & Returns")
plt.legend()
plt.show()

# Pie Chart for Investments
plt.figure(figsize=(6, 6))
plt.pie(investments, labels=stocks, autopct="%1.1f%%", startangle=140, colors=['b', 'g', 'r', 'c', 'm'])
plt.title("Investment Distribution among Stocks")
plt.show()
```
