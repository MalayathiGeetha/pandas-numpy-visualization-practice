```python
import matplotlib.pyplot as plt

# Sales data
days = range(1, 8)  # Days from 1 to 7
week1_sales = [5000, 4000, 4000, 5900, 6500, 3500, 4000]
week2_sales = [4000, 3000, 5800, 5000, 5500, 3000, 4300]
week3_sales = [4000, 5800, 3500, 2500, 3000, 5300, 6000]

# Plotting the data
plt.figure(figsize=(10, 5))
plt.plot(days, week1_sales, marker='*', linestyle='-', color='red', label="Week 1")
plt.plot(days, week2_sales, marker='h', linestyle='-', color='blue', label="Week 2")
plt.plot(days, week3_sales, marker='+', linestyle='-', color='brown', label="Week 3")

# Adding title and labels
plt.title("Mela Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales in Rs")
plt.xticks(days)  # Ensure days are correctly labeled

# Adding legend
plt.legend()

# Show the plot
plt.show()
```
