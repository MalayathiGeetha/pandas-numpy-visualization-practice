```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the waffle chart
data = {'Category A': 30, 'Category B': 20, 'Category C': 10, 'Category D': 40}
total = sum(data.values())
rows = 10
columns = total // rows

# Create the waffle chart grid
waffle_chart = np.zeros((rows, columns))
category_colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Fill the grid with categories
category_index = 0
count = 0

for i in range(rows):
    for j in range(columns):
        if count >= data[list(data.keys())[category_index]]:
            category_index += 1
            count = 0
        waffle_chart[i, j] = category_index
        count += 1

# Plotting the waffle chart
plt.figure(figsize=(10, 6))
plt.matshow(waffle_chart, cmap=plt.cm.get_cmap('Pastel1', len(data)), fignum=1)
plt.xticks([])
plt.yticks([])

# Create legend
legend_labels = [f"{category} ({value}%)" for category, value in data.items()]
patches = [plt.Line2D([0], [0], marker='s', color='w', markerfacecolor=color, markersize=15) 
           for color in category_colors]
plt.legend(patches, legend_labels, loc='lower right', bbox_to_anchor=(1, 0))

plt.title('Waffle Chart Example')
plt.show()
```

```python
# Data
data = {'Apples': 35, 'Bananas': 25, 'Oranges': 20, 'Grapes': 20}
rows = 10
total = sum(data.values())
columns = total // rows

# Create grid
waffle_chart = np.zeros((rows, columns))
categories = list(data.keys())
values = list(data.values())

# Fill the chart
cell = 0
for i, value in enumerate(values):
    for _ in range(value):
        row, col = divmod(cell, columns)
        waffle_chart[row, col] = i
        cell += 1

# Plot
plt.matshow(waffle_chart, cmap='Set3')
plt.xticks([])
plt.yticks([])
plt.legend(categories, loc='lower right')
plt.title('Fruit Waffle Chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Animal data
data = {'Dogs': 40, 'Cats': 30, 'Birds': 20, 'Fish': 10}
total = sum(data.values())
rows = 10
columns = total // rows

# Create a grid of circles
x, y = np.meshgrid(np.arange(columns), np.arange(rows))
category_index = 0
counter = 0

# Assign categories to each circle
colors = sns.color_palette('Set2', len(data))
plt.figure(figsize=(10, 6))

for i in range(rows):
    for j in range(columns):
        if counter >= list(data.values())[category_index]:
            category_index += 1
            counter = 0
        plt.scatter(j, rows - i - 1, s=500, color=colors[category_index])
        counter += 1

# Add legend and title
plt.xticks([])
plt.yticks([])
plt.legend(data.keys(), loc='lower right')
plt.title('Animal Waffle Chart (Circle Style)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Fruit data
fruit_data = {'Apples': 30, 'Bananas': 25, 'Mangoes': 25, 'Oranges': 20}
total = sum(fruit_data.values())
rows, columns = 10, total // 10

# Generate waffle grid
waffle_chart = np.zeros((rows, columns))
category_index = 0
counter = 0

for i in range(rows):
    for j in range(columns):
        if counter >= list(fruit_data.values())[category_index]:
            category_index += 1
            counter = 0
        waffle_chart[i, j] = category_index
        counter += 1

# Heatmap style
plt.figure(figsize=(10, 6))
sns.heatmap(waffle_chart, linewidths=0.5, linecolor='black', cmap='YlGnBu', cbar=False)

plt.xticks([])
plt.yticks([])
plt.legend(fruit_data.keys(), loc='lower right')
plt.title('Fruit Waffle Chart (Heatmap Style)')
plt.show()
```
