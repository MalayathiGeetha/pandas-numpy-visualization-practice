```python
from wordcloud import WordCloud
txt="python is a widely used high level programming langiage for "
a=WordCloud(width=800,height=400,background_color="white").generate(txt)
plt.imshow(a)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
x_val=[1,2,3,4,5,6,7,8]
y_val=[2,4,6,8,10,3,5,1]
sns.lmplot(x='x',y='y',data=pd.DataFrame({'x':x_val,'y':y_val}))
plt.show()
```
