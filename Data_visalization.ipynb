```python
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,10,0.1)
y=2*x+5
plt.plot(x,y,color='pink')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title("Line plot")
plt.show()
```

```python
x=np.arange(1,10,0.1)
y1=2*x+5
y2=3*x+10
plt.subplot(1,2,1)
plt.plot(x,y1)
plt.subplot(1,2,2)
plt.plot(x,y2)
plt.show()
```

```python
#bar plot
fruit={'apple':30,'mango':45,'banana':10}
names=list(fruit.keys())
quan=list(fruit.values())
plt.bar(names,quan,color="orange")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#scatter-plot
x=[10,20,30,40,50,60,70,80,90]
a=[1,2,3,4,5,4,3,2,9]
b=[5,6,7,8,3,2,1,4,9]
plt.figure(figsize=(7,7))
plt.scatter(x,a)
plt.scatter(x,b,s=500,marker='4')
plt.legend(['a','b'])
plt.title('X Vs a&b')

plt.xlabel('x')
plt.ylabel('a&b')
plt.show()
```

```python
#histogram
data=[1,1,1,5,6,3,0,2,7,3,9,1,7,5,4,3,1,1,5]
plt.hist(data,bins=50)
plt.show()
```

```python
#boxplot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data)
plt.show()
```

```python
#pie chart
fruit=['apple','banana','orange','grapes']
quan=[30,40,12,100]
plt.pie(quan,labels=fruit)
plt.show()
```

```python
#pie chart
fruit=['apple','banana','orange','grapes']
quan=[30,40,12,100]
plt.pie(quan,labels=fruit,autopct='%0.1f%%',shadow=True,explode=(0,0,0,0.1),colors=['black','gray','yellow','green'])
plt.show()
```

```python
fruit=['apple','banana','orange','grapes']
quan=[30,40,12,100]
plt.pie(quan,labels=fruit,autopct='%0.1f%%',explode=(0,0.1,0,0.1),colors=['black','gray','yellow','green'])
plt.show()
```

```python
fruit=['apple','banana','orange','grapes']
quan=[30,40,12,100]
plt.pie(quan,labels=fruit,autopct='%0.1f%%',radius=2)
plt.pie([5],colors='w',radius=1)
plt.show()
```

```python
#areaplot 
x=[10,20,30,40,50,60,70,80,90]
y=[1,1,2,5,6,4,7,4,6]
plt.stackplot(x,y,color='green')
plt.plot(x,y,color='red')  #for outline
plt.show()
```
