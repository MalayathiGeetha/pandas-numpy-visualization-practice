```python
#lineplot
from matplotlib import pyplot as plt
import numpy as np
x=np.arange(1,11)
y1=2*x
y2=3*x

plt.plot(x,y1,color='r',linestyle=":")
plt.plot(x,y2,color='y',linewidth='2',linestyle='-.')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line PLot')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=":")
plt.subplot(1,2,2)
plt.plot(x,y2,color='g',linewidth='2',linestyle='-.')
plt.show()
```

```python
#bar plot 
student={"Bob":87,"Matt":56,"Sam":27}
names=list(student.keys())
values=list(student.values())
plt.bar(names,values)
plt.show()
```
